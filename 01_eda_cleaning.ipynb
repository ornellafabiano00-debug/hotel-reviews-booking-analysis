Analisi dei valori mancanti effettuata in precedenza:

"**Gestione "missing values":**<br>

Prima di procedere alla rimozione dei record contenenti valori mancanti (NaN) nelle colonne 'lat' e 'lng',<br>è fondamentale verificare se l'assenza di tali dati sia puramente casuale oppure se risponda a un pattern sistematico.

Se i valori mancanti fossero concentrati interamente su specifici hotel, la loro cancellazione introdurrebbe un bias geografico,<br> eliminando completamente quelle strutture dalle successive analisi.

```python
# Numero totale di hotel unici nel dataset
# Usiamo la chiave composta (Hotel_Name + Hotel_Address) invece del solo nome,
# perché un nome può essere condiviso da hotel diversi in città diverse
# (es. "Hotel Regina" esiste a Barcelona, Vienna e Milano: 3 hotel indipendenti,
# non un duplicato dello stesso hotel)

totale_hotel_per_nome = df_original['Hotel_Name'].nunique()
totale_hotel_reali = df_original.drop_duplicates(subset=['Hotel_Name', 'Hotel_Address']).shape[0]

print(f"Nomi hotel unici (solo Hotel_Name): {totale_hotel_per_nome}")
print(f"Hotel realmente distinti (Hotel_Name + Hotel_Address): {totale_hotel_reali}")
```

```python
# Hotel coinvolti dai valori mancanti
righe_senza_geo = df_original[df_original['lat'].isnull()]
elenco_hotel_mancanti = righe_senza_geo['Hotel_Name'].value_counts().reset_index()
elenco_hotel_mancanti.index = elenco_hotel_mancanti.index + 1

#Risultato finale con il conteggio numerico 
elenco_hotel_mancanti
```

```python
# Hotel coinvolti dai valori mancanti e verifica delle Coordinate_Presenti, Coordinante_Mancanti e Recensioni_Totali.
nomi_hotel_mancanti = df_original[df_original['lat'].isnull()]['Hotel_Name'].unique()

df_filtrato = df_original[df_original['Hotel_Name'].isin(nomi_hotel_mancanti)]

conteggio_geo = df_filtrato.groupby('Hotel_Name').agg(
    Coordinate_Presenti=('lat', lambda x: x.notnull().sum()),
    Coordinate_Mancanti=('lat', lambda x: x.isnull().sum()),
    Recensioni_Totali=('Hotel_Name', 'count')
).reset_index()

conteggio_geo = conteggio_geo.sort_values(by='Coordinate_Mancanti', ascending=False).reset_index(drop=True)

conteggio_geo.index = conteggio_geo.index + 1

#Tabella finale
conteggio_geo
```

Analisi dei Valori mancanti e del Bias Geografico:

Dall'analisi esplorativa iniziale è emersa l'assenza di coordinate geografiche (`lat` e `lng`) per **3.268 record**,<br>pari a circa lo **0,63%** dell'intero dataset (515K righe). 

Prima di procedere a una rimozione tramite `.dropna()`, è stata effettuata una verifica approfondita per capire la natura di questi dati mancanti.
L'obiettivo era determinare se i dati mancassero in modo del tutto casuale o secondo un pattern sistematico legato a specifiche strutture.

L'analisi aggregata ha rivelato una situazione molto chiara e critica:
* **Nomi hotel nel dataset:** Il dataset contiene **1.492 nomi hotel distinti** (attenzione: non equivale al numero di hotel realmente distinti; un caso emerso più avanti nel progetto, "Hotel Regina", è un nome condiviso da 3 hotel indipendenti in città diverse e il conteggio reale di hotel univoci, verificato nel database, è 1.494).
* **I valori mancanti non sono casuali:** I 3.268 valori mancanti sono concentrati interamente ed esclusivamente su **appena 17 hotel**<br>(l'1,1% delle strutture totali).
* **Mancanza totale:** Per ciascuno di questi 17 hotel, la colonna `Coordinate_Presenti` è pari a **0**.<br>
Questo significa che per queste specifiche strutture non esiste alcuna coordinata geografica in nessuna delle loro recensioni nel dataset originale.


**Conclusione:**<br>
Eliminare queste righe con un semplice `.dropna()` introdurrebbe un **bias geografico sistematico**, poiché significherebbe cancellare
non un campione casuale di recensioni, ma tutte le recensioni relative a 17 hotel specifici: si eliminerebbero così
sistematicamente quelle strutture (e le città a cui appartengono) dall'analisi, invece di perdere un dato isolato e ininfluente.

```python
# Reviewer_Score in 4 fasce

# Aggiornamento:
# right=False rende gli intervalli chiusi a sinistra [0,5), [5,7), [7,9), [9,10.1)
# per allinearsi esattamente ai confini usati nella query 
# SQL 2.10 Distribuzione lunghezza recensione per categoria di score (WHEN score < 5, < 7, < 9)
# Nota: l'ultimo bin arriva a 10.1 perché con right=False l'intervallo [9,10) escluderebbe
# il valore massimo 10.0, che deve invece ricadere in "Ottimo"

df['Score_Category'] = pandas.cut(
    df['Reviewer_Score'],
    bins=[0, 5, 7, 9, 10.1],
    labels=['Basso (0-5)', 'Medio (5-7)', 'Buono (7-9)', 'Ottimo (9-10)'],
    right=False
)

# Verifica

# Aggiornamento
# Verifica: i conteggi devono coincidere con la query SQL 2.10
# (attesi: Basso=22232, Medio=64485, Buono=181668, Ottimo=246827)
print(df['Score_Category'].value_counts().sort_index())
```

```python
# 1. Dizionario Nazione -> Città
mappa_geografica = {
    'United Kingdom': 'London',
    'France': 'Paris',
    'Netherlands': 'Amsterdam',
    'Spain': 'Barcelona',
    'Italy': 'Milan',
    'Austria': 'Vienna'
}

# 2. Funzione per assegnare la città in base al suffisso geografico dell'indirizzo
def assegna_citta(indirizzo):
    indirizzo_str = str(indirizzo).strip()
    for nazione, citta in mappa_geografica.items():
        if indirizzo_str.endswith(nazione):
            return citta
    return 'Unknown'

# 3. Applicazione della funzione per creare la variabile derivata richiesta
df['City'] = df['Hotel_Address'].apply(assegna_citta)

# 4. Verifica della distribuzione finale e controllo assenza di 'Unknown'
print(df['City'].value_counts())
print("\nRighe non riconosciute:", (df['City'] == 'Unknown').sum())
```
